A Layer of Neurons

```python
# Doing with Numpy
import numpy as np
```

```python
#List of inputs

inputs = [[2,4,5,8],[2,3,2,1],[2,44,5,6],[1,2,3,4]]
inputs = np.array(inputs , dtype = np.float64)
inputs.shape
```

```python
# List of weights

weights= [[-0.3,0.3,0.2],[0.2,0.4,.9],[.33,.54,.45],[-.12,-.11,.12]]
weights = np.array(weights , dtype=np.float64)
weights.shape
```

```python
# Output of layer

layer_outputs = []
```

```python
for neuron_weights , neuron_bias in zip(weights,biases):
  # Current neuron output
  neuron_output = 0
  # for each input and weights in the neuron

  for n_input , weight in zip(inputs,neuron_weights):
    #Multiplying adding with current weights and input
    neuron_output= neuron_output + n_input*weight

  #Adding the biases
  neuron_output+=neuron_bias
  #Put neuron's result to layer_outputs
  layer_outputs.append(neuron_output)
```

```python
import nnfs
import matplotlib.pyplot as plt
import numpy as np
```

```python
from nnfs.datasets import spiral_data
```

```python
nnfs.init()
```

```python
# Dense Layer

class Layer_Dense:

  #Layer Initialization
  def __init__(self,n_inputs,n_neurons):

    # initializing weights and biases
    self.weights = 0.01 * np.random.rand(n_inputs,n_neurons)
    self.biases = np.zeros((1 , n_neurons))

  # Forward Pass
  def forward(self,inputs):

    #Calculate output from input,weight and biases
    self.output = np.dot(inputs , self.weights) + self.biases
```

```python
#create Dataset

X,y = spiral_data(samples=100,classes=3)
```

```python
# Create a Dense layer


dense = Layer_Dense(2 , 3)
```

```python
# Perform a Forward Pass of the training data through this layer

dense.forward(X)
```

```python
import numpy as np
import nnfs
from nnfs.datasets import spiral_data
```

```python
# ReLu Activation

class Activation_ReLu:

  # Forward Pass

  def forward(self , inputs):

    # Calculate ouptut from input

    self.outputs = np.maximum(0,inputs)
```

```python
# Create Dataset

X , y = spiral_data(samples=100,classes=3)
```

```python
# Dense Layer

dense = Layer_Dense(2,3)
```

```python
# Create ReLu Activation

activation = Activation_ReLu()
```

```python
# Making  a forward pass to our training data through the dense layer

dense.forward(X)
```

```python
# Forward pass through activation function
# Take input from previous output

activation.forward(dense.output)
```

```python
# Print the output

print(activation.outputs)

```

N.B -> In 2d array axis = 0 refers to the axis and axis = 1 refers to the column

Keepdims is a boolean type parameter, if this is set to True, the axes which are reduced are left in the result as dimensions with size one. With this option, the result will broadcast correctly against the input array.

```python
# Softmax Activation

class Activation_Softmax:

  # Forward Pass

  def forward(self , inputs):

    # Get Unnormalized Probabilites

    exp_values = np.exp(inputs - np.max(inputs , axis = 1 , keepdims=True))

    # Normalize them for each sample

    probabilites = exp_values / np.sum(exp_values , axis = 1 , keepdims=True)
    self.output =  probabilites
```

```python
# create a Dataset
X , y = spiral_data(samples=100 , classes = 3)
```

```python
# dense layer with 2 input features and 3 output values

dense1 = Layer_Dense(2 , 3)
```

```python
# Create ReLu  activation function

activation = Activation_ReLu()
```

```python
# Create a second layer with 3 input features (as we take 3 ouput features from dense1) ans 3 output values

dense2 = Layer_Dense(3 , 3)
```

```python
# Create Softmax Activation (to be used with Dense Layaer)\

activation2 = Activation_Softmax()
```

```python
# Make a forward pass our training data through this layer

dense1.forward(X)
```

```python
# Make a forward pass through activation function , it takes output of the first dense layer here

activation.forward(dense1.output)
```

```python
# Make a forward pass through our second layer ,,,,, it takes outputs of the first layer activation function as inputs

dense2.forward(activation.outputs)
```

```python
# Make a forward pass through softmax activation ,,,, it takes outputs  of the second dense layer funtion as inputs

activation2.forward(dense2.output)
```

```python
# Print the values of activaton function of second layer

print(activation2.output[:5])
```
